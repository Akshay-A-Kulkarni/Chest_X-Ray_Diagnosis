```python
from x_ray import preprocessor
import sklearn.metrics
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelBinarizer
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Data Import

```python
csv_path = '/Users/gk/Desktop/sample/x_ray_1000.csv'
image_path = '/Users/gk/Desktop/sample/x_ray_1000_images/'
```

# Binary Classification

```python
report_labels = ['Cardiomegaly','Consolidation','Effusion','Mass']

for label in report_labels:

    input = ['No Finding',label]
    images = preprocessor(csv_path, image_path, 100, 100, input)
    X_train, X_val, X_test, y_train, y_val, y_test = images.prepare_x_ray_data()
    logistic_model = LogisticRegression(solver = 'lbfgs',multi_class='multinomial',max_iter=500)
    logistic_model = logistic_model.fit(X_train,y_train)
    logistic_pred = logistic_model.predict(X_test)
    logistic_accuracy = accuracy_score(y_test,logistic_pred)
    print(label)
    print("Accuracy: ",logistic_accuracy)
    logistic_precision = sklearn.metrics.precision_score(y_test,logistic_pred,average='macro')
    print("Precision: ",logistic_precision)
    logistic_recall = sklearn.metrics.recall_score(y_test,logistic_pred,average='macro')
    print("Recall: ",logistic_recall)
    logistic_f1 = sklearn.metrics.f1_score(y_test,logistic_pred,average='macro')
    print("F1: ",logistic_f1)
    print("AUC: ",roc_auc_score(y_test, logistic_pred, average="macro"))
    cm = sklearn.metrics.confusion_matrix(y_test, logistic_pred)
    plt.figure(figsize=(2,2))
    sns.heatmap(cm, annot=True, fmt=".2f", linewidths=0.5, square = False, cmap = 'Blues_r', xticklabels=True,yticklabels=True)
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.xlim([2.0,0.0])
    plt.ylim([0.0,2.0])
```

# Multi-Classification

```python
input_labels = ['Atelectasis','Cardiomegaly','Consolidation','Effusion','Infiltration','Mass',
               'No Finding', 'Nodule', 'Pleural_Thickening', 'Pneumothorax']

images = preprocessor(csv_path, image_path, 100, 100, input_labels)
X_train, X_val, X_test, y_train, y_val, y_test = images.prepare_x_ray_data()
logistic_model = LogisticRegression(solver = 'lbfgs',multi_class='ovr',max_iter=500)
logistic_model = logistic_model.fit(X_train,y_train)
logistic_pred = logistic_model.predict(X_test)
logistic_accuracy = accuracy_score(y_test,logistic_pred)
print("Accuracy: ",logistic_accuracy)
logistic_precision = sklearn.metrics.precision_score(y_test,logistic_pred,average='macro')
print("Precision: ",logistic_precision)
logistic_recall = sklearn.metrics.recall_score(y_test,logistic_pred,average='macro')
print("Recall: ",logistic_recall)
logistic_f1 = sklearn.metrics.f1_score(y_test,logistic_pred,average='macro')
print("F1: ",logistic_f1)
```

```python
def multiclass_roc_auc_score(y_test, y_pred, average="macro"):
  lb = LabelBinarizer()
  lb.fit(y_test)
  y_test = lb.transform(y_test)
  y_pred = lb.transform(y_pred)
  return roc_auc_score(y_test, y_pred, average=average)
print("AUC: ",multiclass_roc_auc_score(y_test, logistic_pred, average="macro"))
```

### ROC Curves, One vs All

```python
def multiclass_roc_curve(y_test, y_pred):
    lb = LabelBinarizer()
    lb.fit(y_test)
    y_test = lb.transform(y_test)
    y_pred = lb.transform(y_pred)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(10):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_pred[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    return fpr, tpr, roc_auc

fpr, tpr, roc_auc = multiclass_roc_curve(y_test, logistic_pred)
plt.figure()
lw = 2
plt.plot(fpr[0], tpr[0], color='darkorange',
         lw=lw, label='Atelectasis (area = %0.2f)' % roc_auc[0])
plt.plot(fpr[1],tpr[1], color='cyan',
         lw=lw, label='Cardiomegaly (area = %0.2f)' % roc_auc[1])
plt.plot(fpr[2],tpr[2],color='red',
         lw=lw, label= 'Consolidation (area = %0.2f)' % roc_auc[2])
plt.plot(fpr[3],tpr[3],color='pink',
         lw=lw, label = 'Effusion (area = %0.2f)' % roc_auc[3])
plt.plot(fpr[4],tpr[4],color='green',
         lw=lw, label = 'Infiltration (area = %0.2f)' % roc_auc[4])
plt.plot(fpr[5],tpr[5],color='yellow',
         lw=lw, label = 'Mass (area = %0.2f)' % roc_auc[5])
plt.plot(fpr[6],tpr[6],color='indigo',
         lw=lw, label = 'No Finding (area = %0.2f)' % roc_auc[6])
plt.plot(fpr[7],tpr[7],color='blue',
         lw=lw, label = 'Nodule (area = %0.2f)' % roc_auc[7])
plt.plot(fpr[8],tpr[8],color='skyblue',
         lw=lw, label = 'Pleural_Thickening (area = %0.2f)' % roc_auc[8])
plt.plot(fpr[9],tpr[9],color='forestgreen',
         lw=lw, label = 'Pneumothorax (area = %0.2f)' % roc_auc[9])
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

### Confusion Matrix

```python
cm = sklearn.metrics.confusion_matrix(y_test, logistic_pred)
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True, fmt=".2f", linewidths=0.5, square = False, cmap = 'Blues_r', xticklabels=True,yticklabels=True)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix',size = 15)
plt.xlim([0.0,10.0])
plt.ylim([10.0,0.0])
```
